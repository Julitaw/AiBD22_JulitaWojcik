```python
import pandas as pd
data = pd.read_csv('../OriginalData/tb.csv')
data
```

```python
data_without_null = data.dropna(axis=0,how='all', thresh=21)
data_without_null
data_without_null.fillna(0)
```

```python
df = data_without_null.fillna(0)
#df = (pd.read_csv('../OriginalData/tb.csv')).dropna()
df = pd.melt(df, id_vars=['iso2', 'year'], value_vars=list(df.columns)[3:],
             var_name='column', value_name='cases')
df = df[df['cases'] != '0']
df['cases'] = df['cases'].astype(int)
df['sex'] = df['column'].str[7]
df['age'] = df['column'].str[8:].map({
    '04': '0-4',
    '514': '5-14',
    '014': '0-14',
    '1524': '15-24',
    '2534': '25-34',
    '3544': '35-44',
    '4554': '45-54',
    '5564': '55-64',
    '65': '65+',
    'u': 'unknown'
})
df = df[['iso2', 'year', 'sex', 'age', 'cases']]
#df.to_csv('../Originaldata/tb-tidy.csv', index=False)
x = df.loc[(df['cases'] != 0)]
x.sort_values(by=['year'])
x = df.loc[(df['iso2'] == 'CL')]
x.reset_index(drop=True)

```
